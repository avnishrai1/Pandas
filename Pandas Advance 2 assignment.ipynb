```python
"""
Q2. What is the difference between the functions loc and iloc in pandas.DataFrame?
Ans.
"""
import pandas as pd

course_name = ['Data Science', 'Machine Learning', 'Big Data', 'Data Engineer']

duration = [2,3,6,4]

df = pd.DataFrame(data = {'course_name' : course_name, 'duration' : duration})

print(df.loc[1:2])
print(df.iloc[1])
```

```python
"""
Q3. Reindex the given dataframe using a variable, reindex = [3,0,1,2] and store it in the variable, new_df then find the output for
    both new_df.loc[2] and new_df.iloc[2].
    Did you observe any difference in both the outputs? If so then explain it.
Ans.
    Original DataFrame df:
       course_name  duration
0  Data Science         2
1  Machine Learning   3
2  Big Data          6
3  Data Engineer     4

    After Reindex :
   course_name  duration
3  Data Engineer     4
0  Data Science         2
1  Machine Learning   3
2  Big Data          6

The reindex list [3, 0, 1, 2] specifies the new order in which the rows should appear in the new_df. It effectively says that the row with
index 3 in the original DataFrame should become the first row, the row with index 0 should become the second row, and so on.

The df.reindex(reindex) method is called, which creates a new DataFrame new_df with the rows reordered based on the reindex

new_df.loc[2] will use label-based indexing (loc) to select the row with label 2. In this reordered DataFrame, the row with 
label 2 corresponds to 'Big Data' with a duration of 6. So, new_df.loc[2]

new_df.iloc[2] will use integer-based indexing (iloc) to select the row at position 2. In this reordered DataFrame, the row at position 
2 corresponds to 'Machine Learning' with a duration of 3.
"""
import pandas as pd

course_name = ['Data Science', 'Machine Learning', 'Big Data', 'Data Engineer']

duration = [2,3,6,4]
reindex = [3,0,1,2]

df = pd.DataFrame(data = {'course_name' : course_name, 'duration' : duration})

new_df = df.reindex(reindex)

print(new_df.loc[2])
print(new_df.iloc[2])
```

```python
"""
Q6. What do you understand about the windows function in pandas and list the types of windows
    functions?
Ans.In pandas, a window function, also known as a rolling or moving window function, is a powerful tool for performing calculations on 
    a sliding window of data within a DataFrame or Series. It's particularly useful for time-series data or any data where you want to 
    analyze trends, moving averages, or other aggregations over a specific window or interval of data points.
    
    Window functions in pandas allow you to perform operations like:
    Rolling Mean: Calculating the mean of data points within a rolling window.
    Rolling Sum: Calculating the sum of data points within a rolling window.
    Rolling Standard Deviation: Calculating the standard deviation of data points within a rolling window.
    Rolling Min and Max: Finding the minimum and maximum values within a rolling window.
    Exponential Moving Average (EMA): Calculating the exponential moving average, which gives more weight to recent data points.
"""
import pandas as pd

data = {'value': [10, 20, 30, 40, 50]}
df = pd.DataFrame(data)


rolling_window = df['value'].rolling(window=3)

print(rolling_window.mean())
print(rolling_window.sum())
print(rolling_window.std())
print(rolling_window.min())
print(rolling_window.max())
```

```python
"""
Q8. Write a Python program that takes in two dates as input (in the format YYYY-MM-DD) and
calculates the difference between them in days, hours, and minutes using Pandas time delta. The
program should prompt the user to enter the dates and display the result.
"""
import pandas as pd

date_str1 = input("Enter the first date (YYYY-MM-DD): ")

date_str2 = input("Enter the second date (YYYY-MM-DD): ")

try:
    
    date1 = pd.to_datetime(date_str1)
    date2 = pd.to_datetime(date_str2)

    time_difference = date2 - date1

    days = time_difference.days
    seconds = time_difference.seconds
    hours, seconds = divmod(seconds, 3600)
    minutes, seconds = divmod(seconds, 60)

    print(f"Time difference: {days} days, {hours} hours, {minutes} minutes")
except ValueError:
    print("Invalid date format. Please use the format 'YYYY-MM-DD'.")
```

```python
"""
Q9. Write a Python program that reads a CSV file containing categorical data and converts a specified column to a categorical data 
    type. The program should prompt the user to enter the file path, column name, and category order, and then display the sorted data.
"""
import pandas as pd

# Prompt the user to enter the file path
file_path = input("Enter the CSV file path: ")

# Prompt the user to enter the column name
column_name = input("Enter the column name to convert to categorical: ")

# Prompt the user to enter the custom category order as a comma-separated list
category_order = input("Enter the custom category order (comma-separated): ").split(',')

try:
    # Read the CSV file into a DataFrame
    df = pd.read_csv(file_path)

    # Convert the specified column to a categorical data type with custom categories
    df[column_name] = pd.Categorical(df[column_name], categories=category_order, ordered=True)

    # Sort the DataFrame by the specified column
    sorted_df = df.sort_values(by=column_name)

    # Display the sorted DataFrame
    print(sorted_df)
except FileNotFoundError:
    print("File not found. Please provide a valid file path.")
except KeyError:
    print(f"Column '{column_name}' not found in the CSV file.")
```
